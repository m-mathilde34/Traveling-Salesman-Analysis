```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_to_data = "../data/"
```

```python
filename = "7_GA_MediumCities_007_Results.csv"

dataframe = pd.read_csv(path_to_data + filename)
dataframe.head()
```

```python
filename2 = "8_GA_MediumCities_008_Results.csv"

dataframe2 = pd.read_csv(path_to_data + filename2)
dataframe2.head()
```

```python
filename3 = "9_GA_MediumCities_009_Results.csv"

dataframe3 = pd.read_csv(path_to_data + filename3)
dataframe3.head()
```

```python
filename4 = "10_GA_MediumCities_010_Results.csv"

dataframe4 = pd.read_csv(path_to_data + filename4)
dataframe4.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

sns.lineplot(data=dataframe, x="generation", y="average_fitness", label="1,100,40,0.6,1", ax=axs[0])
sns.lineplot(data=dataframe, x="generation", y="best_fitness", label="1,100,40,0.6,1", ax=axs[1])
sns.lineplot(data=dataframe2, x="generation", y="average_fitness", label="2,100,40,0.1,1", ax=axs[0])
sns.lineplot(data=dataframe2, x="generation", y="best_fitness", label="2,100,40,0.1,1", ax=axs[1])
sns.lineplot(data=dataframe3, x="generation", y="average_fitness", label="3,100,20,0.6,1", ax=axs[0])
sns.lineplot(data=dataframe3, x="generation", y="best_fitness", label="3,100,20,0.6,1", ax=axs[1])
sns.lineplot(data=dataframe4, x="generation", y="average_fitness", label="4,50000,40,0.1,1", ax=axs[0])
sns.lineplot(data=dataframe4, x="generation", y="best_fitness", label="4,50000,40,0.1,1", ax=axs[1])

axs[0].set_title("Average Fitness")
axs[1].set_title("Best Fitness")
plt.legend()
plt.show()
```
